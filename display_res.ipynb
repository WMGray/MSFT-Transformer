```python
def get_same_paras_results(path: str, groupby_features) -> pd.DataFrame:
    df = pd.read_csv(path)
    group_df = df.groupby(groupby_features)
    avg_results = []
    for name, tmp_df in group_df:
        if tmp_df.shape[0] != 5:
            continue
        columns = list(tmp_df.columns)
        average_metrics = {k: tmp_df[k].iloc[0] for k in columns if k not in ['AUC', 'ACC', 'Recall', 'Precision', 'F1']}
        average_metrics.update({
            'Average AUC': round(tmp_df['AUC'].mean(), 4) * 100,
            'Average AUC std': round(sem(tmp_df['AUC']), 5) * 100,
            'Average ACC': round(tmp_df['ACC'].mean(), 4) * 100,
            'Average Recall': round(tmp_df['Recall'].mean(), 4) * 100,
            'Average Precision': round(tmp_df['Precision'].mean(), 4) * 100,
            'Average F1': round(tmp_df['F1'].mean(), 4) * 100,
        })
        avg_results.append(average_metrics)
    result_df = pd.DataFrame(avg_results)
    result_df =result_df.sort_values(by='Average AUC', ascending=False)
    return result_df

```

```python
disease = "C-T2DD"
path = f"./results/{disease}/MTMFTransformer_evaluate.csv"
group_feats = ['n_layers', 'num_bottleneck', 'lr', 'batch_size', 'use_bottleneck', 'btn_init', 'use_cross_atn', 'mode']  
# group_feats = ['n_blocks', 'lr', 'batch_size', 'mode']  FT-Transformer
# group_feats = ['n_layers', 'm_layers', 'num_bottleneck', 'hidden_size', 'lr', 'batch_size', 'mode']
result_df = get_same_paras_results(path, group_feats)
result_df.head(5)
```
